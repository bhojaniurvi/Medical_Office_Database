```python
# import required libraries
import pymysql

# connect to MySQL server and database
cnx = pymysql.connect(host='localhost', user='root', password='2016@Texas', db='doc_office',
                      cursorclass=pymysql.cursors.DictCursor)

# Create a cursor
cursor = cnx.cursor()

# Department Functions
def read_department():
    try:
       cursor.callproc('readDepartment')
       rows = cursor.fetchall()
       for row in rows:
           print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")
        
# Physician Functions
def create_physician():
    try:
        physician_name = input("Enter the physician name: ")
        physician_type = input("Enter the physician's department: ")
        cursor.callproc('createPhysician', (physician_name, physician_type))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")
        
def delete_physician():
    try:
        physician_id = int(input("Enter the physician ID to delete: "))
        cursor.callproc('removePhysician', (physician_id,))
        cnx.commit()
        print("Record deleted successfully.")
    except pymysql.Error as e:
        print(f"Error deleting the record: {e}")

def read_physician():
    try:
       cursor.callproc('readPhysician')
       rows = cursor.fetchall()
       for row in rows:
           print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

def fetch_physician():
    try:
       physician_id = int(input("Enter the physician ID to fetch: "))
       cursor.callproc('fetchPhysician', (physician_id,))
       rows = cursor.fetchall()
       print(rows)
    except pymysql.Error as e:
        print(f"Error fetching the record: {e}")

# Insurance Functions
def create_ins():
    try:
        company_name = input("Enter the insurance company's name: ")
        type = input("Enter the type of insurance: ")
        cursor.callproc('createInsuranceType', (company_name, type))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def read_ins():
    try:
        cursor.callproc('readInsurance')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

# Patient Functions
def create_patient():
    try:
        ssn = input("Enter the patient's social security number: ")
        patient_name = input("Enter the patient's name: ")
        address = input("Enter the patient's address: ")
        phone = input("Enter the patient's phone number: ")
        insurance_id = input("Enter the patient's insurance id: ")
        cursor.callproc('createPatient', (ssn, patient_name, address, phone, insurance_id))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def delete_patient():
    try:
        patientId = int(input("Enter patient ID to delete: "))
        cursor.callproc('removePatient', (patientId,))
        cnx.commit()
        print("Record deleted successfully.")
    except pymysql.Error as e:
        print(f"Error deleting the record: {e}")

def read_patient():
    try:
        cursor.callproc('readPatient')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

def update_patient():
    try:
        p_patient_id = input("Enter the patient ID to update: ")
        p_ssn = input("Enter the ssn to update: ")
        p_patient_name = input("Enter the patient name to update: ")
        p_address = input("Enter the address to update: ")
        p_phone = input("Enter the phone number to update: ")
        p_insurance_id = input("Enter the insurance ID to update: ")
        cursor.callproc('updatePatient', (p_patient_id, p_ssn, p_patient_name, p_address, p_phone, p_insurance_id))
        cnx.commit()
        print("Record updated successfully.")
    except pymysql.Error as e:
        print(f"Error updating the record: {e}")
        
def fetch_patient():
    try:
       patient_id = int(input("Enter the patient ID to fetch: "))
       cursor.callproc('fetchPatient', (patient_id,))
       rows = cursor.fetchall()
       print(rows)
    except pymysql.Error as e:
        print(f"Error fetching the record: {e}")

# Equipment Functions
def create_equipment():
    try:
        equip_name = input("Enter the name of the piece of equipment: ")
        equip_desc = input("Enter a description of the piece of equipment: ")
        cursor.callproc('addEquipment', (equip_name, equip_desc))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def delete_equipment():
    try:
        p_equip_id = int(input("Enter equipment ID to delete: "))
        cursor.callproc('removeEquipment', (p_equip_id,))
        cnx.commit()
        print("Record deleted successfully.")
    except pymysql.Error as e:
        print(f"Error deleting the record: {e}")

def read_equipment():
    try:
        cursor.callproc('readEquipment')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

# Medication Functions
def create_med():
    try:
        med_name = input("Enter the name of the medication: ")
        med_desc = input("Enter a description of the medication: ")
        cursor.callproc('addEquipment', (med_name, med_desc))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def read_med():
    try:
        cursor.callproc('readMedication')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

def fetch_medication():
    try:
       med_id = int(input("Enter the medication ID to fetch: "))
       cursor.callproc('fetchMedication', (med_id,))
       rows = cursor.fetchall()
       print(rows)
    except pymysql.Error as e:
        print(f"Error fetching the record: {e}")
        
# Prescription Functions
def create_pres():
    try:
        pat = input("Enter the patient's id: ")
        phys = input("Enter the physician's id : ")
        med = input("Enter the medicine id: ")
        cursor.callproc('createPrescription', (pat, phys, med))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def update_pres():
    try:
        p_presc_id = input("Enter the prescription id to update: ")
        p_pat = input("Enter the patient's id to update: ")
        p_phys = input("Enter the physician's id to update: ")
        p_med = input("Enter the medication id: ")
        cursor.callproc('updatePrescription', (p_presc_id, p_pat, p_phys, p_med ))
        cnx.commit()
        print("Record updated successfully.")
    except pymysql.Error as e:
        print(f"Error updating the record: {e}")

def read_pres():
    try:
        cursor.callproc('readPrescription')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

# Appointment Functions
def create_appt():
    try:
        p_patient_id = input("Enter the patient's id: ")
        p_physician_id = input("Enter the physician's id : ")
        p_room = input("Enter the room number: ")
        p_date = input("Enter the appointment date (ex: 2023-06-21): ")
        p_start_time = input("Enter the start time (ex: 11:00): ")
        p_end_time = input("Enter the end time (ex: 12:00): ")
        p_description = input("Enter an appointment description: ")
        cursor.callproc('createAppointment', (p_patient_id, p_physician_id, p_room, p_date, p_start_time,
                                              p_end_time, p_description))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def delete_appt():
    try:
        p_apt_id = int(input("Enter appointment ID to delete: "))
        cursor.callproc('cancelAppointment', (p_apt_id,))
        cnx.commit()
        print("Record deleted successfully.")
    except pymysql.Error as e:
        print(f"Error deleting the record: {e}")

def update_appt():
    try:
        p_apt_id = input("Enter the appointment ID to update: ")
        p_room = input("Enter the room number to update: ")
        p_appt_date = input("Enter the appointment data to update: ")
        p_start_time = input("Enter the start time to update: ")
        p_end_time = input("Enter the end time to update: ")
        cursor.callproc('updateAppointment', (p_apt_id, p_room, p_appt_date, p_start_time, p_end_time))
        cnx.commit()
        print("Record updated successfully.")
    except pymysql.Error as e:
        print(f"Error updating the record: {e}")
        
def fetch_appt():
    try:
       appt_id = int(input("Enter the Appointment ID to fetch: "))
       cursor.callproc('fetchAppointment', (appt_id,))
       rows = cursor.fetchall()
       print(rows)
    except pymysql.Error as e:
        print(f"Error fetching the record: {e}")

def read_appt():
    try:
        cursor.callproc('readAppointment')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

# Billing Functions
def create_bill():
    try:
        p_apt = input("Enter the appointment id: ")
        p_amount = input("Enter the total amount to pay: ")
        p_paid_status = input("Did the patient complete their payment plan (0 = yes, 1 = no)?: ")
        cursor.callproc('createBilling', (p_apt, p_amount, p_paid_status))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def update_bill():
    try:
        p_apt = input("Enter the appointment ID to update: ")
        p_paid_status = input("Did the patient complete their payment plan (0 = yes, 1 = no)?: ")
        cursor.callproc('updateAppointment', (p_apt, p_paid_status))
        cnx.commit()
        print("Record updated successfully.")
    except pymysql.Error as e:
        print(f"Error updating the record: {e}")

def read_bill():
    try:
        cursor.callproc('readBilling')
        rows = cursor.fetchall()
        for row in rows:
            print(row)
    except pymysql.Error as e:
        print(f"Error reading the record: {e}")

# Appointment Equipment Functions
def create_appt_equip():
    try:
        p_apt = input("Enter the appointment id: ")
        p_equipment = input("Enter the equipment id: ")
        cursor.callproc('addAppointmentEquipment', (p_apt, p_equipment))
        cnx.commit()
        print("Record created successfully.")
    except pymysql.Error as e:
        print(f"Error creating the record: {e}")

def delete_appt_equip():
    try:
        p_apt = int(input("Enter the appointment ID to delete: "))
        p_equipment = int(input("Enter the equipment ID to delete: "))
        cursor.callproc('removeAppointmentEquipment', (p_apt, p_equipment))
        cnx.commit()
        print("Record deleted successfully.")
    except pymysql.Error as e:
        print(f"Error deleting the record: {e}")
```

```python
# Prompt User : Main Menu
while True:
    print("\n--- Hello! Welcome to the CS3200 Hospital. What would you like to look at today? ---")
    print("1. Physician Records")
    print("2. Patient Records")
    print("3. Appointment Details")
    print("4. Equipment Details")
    print("5. Insurance Details")
    print("6. Prescription Details")
    print("7. Medication Details")
    print("8. Department Details")

    menu_choice = input("\nEnter your choice (must be a number from 1-7): ")

    # Physician Records
    if menu_choice == "1":
        print("\nYou will now be redirected to Physician Records.")
        print("What would you like to do?")
        print("1. Enter new Physician Details")
        print("2. Delete old Physician Details")
        print("3. Read Physician Details")
        print("4. Fetch Physician Details")

        phys_choice = input("\nEnter your choice:")

        if phys_choice == "1":
            create_physician()
        elif phys_choice == "2":
            delete_physician()
        elif phys_choice == "3":
            read_physician()
        elif phys_choice == "4":
            fetch_physician()
        else:
            print("Invalid choice. Please try again.")

    # Patient Records
    elif menu_choice == "2":
        print("\nYou will now be redirected to Patient Records.")
        print("What would you like to do?")
        print("1. Enter new Patient Details")
        print("2. Delete old Patient Details")
        print("3. Read Patient Details")
        print("4. Update Patient Details")
        print("5. Fetch Patient Details")

        pat_choice = input("\nEnter your choice:")

        if pat_choice == "1":
            create_patient()
        elif pat_choice == "2":
            delete_patient()
        elif pat_choice == "3":
            read_patient()
        elif pat_choice == "4":
            update_patient()
        elif pat_choice == "5":
            fetch_patient()
        else:
            print("Invalid choice. Please try again.")

    # Appointment and Medical Equipment Details
    elif menu_choice == "3":
        print("\nYou will now be redirected to Appointment Details.")
        print("What would you like to do?")
        print("1. Enter new Appointment Details")
        print("2. Delete old Appointment Details")
        print("3. Update Appointment Details")
        print("4. Read appointment details")
        print("5. Fetch appointment details")

        appt_choice = input("\nEnter your choice:")

        if appt_choice == "1":
            create_appt()
        elif appt_choice == "2":
            delete_appt()
        elif appt_choice == "3":
            print("\nWould you like to add or delete any medical equipment in this appointment?")
            print("1. Add equipment")
            print("2. Delete equipment")
            print("3. Edit other appointment details")
            med_e = input("\nEnter your choice:")
            if med_e == "1":
                create_appt_equip()
            elif med_e == "2":
                delete_appt_equip()
            elif med_e == "3":
                update_appt()
            else:
                print("Invalid choice. Please try again")
        elif appt_choice == "4":
            read_appt()
        elif appt_choice == "5":
            fetch_appt()
        else:
            print("Invalid choice. Please try again.")

    # Equipment Details
    elif menu_choice == "4":
        print("\nYou will now be redirected to Equipment Details.")
        print("What would you like to do?")
        print("1. Enter new Equipment Details")
        print("2. Delete old Equipment Details")
        print("3. Read Equipment Details")

        equip_choice = input("\nEnter your choice:")

        if equip_choice == "1":
            create_equipment()
        elif equip_choice == "2":
            delete_equipment()
        elif equip_choice == "3":
            read_equipment()
        else:
            print("Invalid choice. Please try again.")

    # Insurance Details
    elif menu_choice == "5":
        print("\nYou will now be redirected to Insurance Details.")
        print("What would you like to do?")
        print("1. Enter new Insurance Details")
        print("2. Read Through Insurance Details")

        ins_choice = input("\nEnter your choice:")

        if ins_choice == "1":
            create_ins()
        elif ins_choice == "2":
            read_ins()
        else:
            print("Invalid choice. Please try again.")

    # Prescription Details
    elif menu_choice == "6":
        print("\nYou will now be redirected to Prescription Details.")
        print("What would you like to do?")
        print("1. Enter new Prescription Details")
        print("2. Update Prescription Details")
        print("3. Read Prescription Details")

        ins_choice = input("\nEnter your choice:")

        if ins_choice == "1":
            create_pres()
        elif ins_choice == "2":
            update_pres()
        elif ins_choice == "3":
            read_pres()
        else:
            print("Invalid choice. Please try again.")

    # Medication Details
    elif menu_choice == "7":
        print("\nYou will now be redirected to Medication Details.")
        print("What would you like to do?")
        print("1. Enter new Medication Details")
        print("2. Read Medication Details")
        print("3. Fetch Medication Details")

        ins_choice = input("\nEnter your choice:")

        if ins_choice == "1":
            create_med()
        elif ins_choice == "2":
            read_med()
        elif ins_choice == "3":
            fetch_medication()
        else:
            print("Invalid choice. Please try again.")
    elif menu_choice == "8":
        print("Departments at Hospital:")
        read_department()
        
    else:
        print("Invalid choice. Please try again.")

# Close the cursor and connection
cursor.close()
cnx.close()
```
